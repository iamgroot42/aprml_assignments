```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def read_file(filename):
    accuracies = []
    epochs = []
    with open(filename) as f:
        for row in f:
            lines = row.rstrip().split(']')
            for line in lines:
                if len(line)>0:
                    values = line.strip(',[ ').rstrip().split(',')
                    epochs.append(int(values[1].strip())+1)
                    accuracies.append(float(values[2].strip()))
    return np.array(epochs), np.array(accuracies)
```

```python
et, at = read_file('./train')
ev, av = read_file('./val')
```

```python
plt.plot(et, at, label='Training accuracy')
plt.plot(ev, av, label='Cross validation accuracy')
plt.title('Training/Validation accuracy v/s Epochs')
plt.ylabel('Accuracy')
plt.xlabel('Epoch')
plt.legend()
plt.show()
```
